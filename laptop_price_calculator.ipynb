```python
def get_Os(data):
  if data=='Windows 10' or data == 'Windows 10 S' or data =='Windows 7':
    return 'Windows'
  elif data == 'Mac OS X' or data=='macOS':
    return 'Mac'
  else:
    return 'No Os/Linus/Chrome OS'
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.compose import make_column_transformer
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2)
```

```python
X
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['Company','TypeName','Processor_com','OS','Gpu_Brand']])
```

```python
column_transfer = make_column_transformer((OneHotEncoder(categories=ohe.categories),['Company','TypeName','Processor_com','OS','Gpu_Brand']),
                                          remainder='passthrough')
```

```python
LinearRegression
```

```python
lr=LinearRegression()
```

```python
model=make_pipeline(column_transfer,lr)
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
r2_score(y_test,y_predict)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
KN = KNeighborsRegressor()
```

```python
model1 = make_pipeline(column_transfer,KN)
```

```python
model1.fit(X_train,y_train)
```

```python
y_predict1 = model1.predict(X_test)
```

```python
r2_score(y_test,y_predict1)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RF=RandomForestRegressor()
```

```python
model2 = make_pipeline(column_transfer,RF)
```

```python
model2.fit(X_train,y_train)
```
